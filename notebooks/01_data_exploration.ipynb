## 3. Deal Volume by Year

```python
deals_by_year = df.groupby('year').size().reset_index(name='deal_count')

fig, ax = plt.subplots(figsize=(9, 4))
ax.bar(deals_by_year['year'], deals_by_year['deal_count'],
       color='#2E75B6', edgecolor='white')
ax.set_xlabel('Year'); ax.set_ylabel('Number of Deals')
ax.set_title('Mittelstand M&A Deal Volume by Year', fontweight='bold', color='#1B3A6B')
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()
```
